# Divide and Conquer

## 4-1: Binary Search


```python
def binary_search(a, low, high, key):
    if high < low:
        return -1
    mid = (high + low) // 2
    if key == a[mid]:
        return mid
    if key < a[mid]:
        return binary_search(a, low, mid - 1, key)
    if key > a[mid]:
        return binary_search(a, mid + 1, high, key)


array = input()
n = int(array.split()[0])
an = list(map(int, array.split()[1:]))
keys = input()
k = int(keys.split()[0])
bn = list(map(int, keys.split()[1:]))

for b in bn:
    index = binary_search(an, 0, n - 1, b)
    print(index, end=' ')
```

## 4-3: Majority Element


```python
n = int(input())
seq = [int(i) for i in input().split()]


def divide_func(seq, l, r):
    if l + 1 == r:
        return seq[l]
    elif l + 2 == r:
        return seq[l]
        
    m = (l + r) // 2
    left = divide_func(seq, l, m)
    right = divide_func(seq, m, r)

    c1, c2 = 0, 0

    for i in seq[l:r]:
        if i == left:
            c1 += 1
        elif i == right:
            c2 += 1

    if c1 > (r - l) // 2 and left != -1:
        return left
    elif c2 > (r - l) // 2 and right != -1:
        return right
    else:
        return -1


print(int(divide_func(seq, 0, n) != -1))
```

## 4-5: Inversions


```python
def merge(left, right):
    i, j, inversion_counter = 0, 0, 0
    final = list()
    
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            final.append(left[i])
            i += 1
        else:
            final.append(right[j])
            inversion_counter += len(left) - i
            j += 1

    final += left[i:]
    final += right[j:]

    return final, inversion_counter


def mergesort(arr):
    global tot_count
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2

    left = mergesort(arr[:mid])
    right = mergesort(arr[mid:])

    sorted_arr, temp = merge(left, right)
    tot_count += temp

    return sorted_arr


tot_count = 0
n = int(input())
seq = [int(i) for i in input().split()]
mergesort(seq)

print(tot_count)
```

## 4-6: Organizing a Lottery



```python
master_list = list()
s, p = [int(i) for i in input().split()]

for i in range(s):
    a, b = [int(i) for i in input().split()]
    master_list.append((a, 'l'))
    master_list.append((b, 'r'))

points = input().split()

for i in points:
    master_list.append((int(i), 'p'))

master_list.sort()

segment_count = 0
point_segment_map = dict()

for i in master_list:
    if i[1] == 'l':
        segment_count += 1
    elif i[1] == 'r':
        segment_count -= 1
    else:
        point_segment_map[i[0]] = segment_count
 
temp = ''

for i in points:
    temp += str(point_segment_map[int(i)]) + ' '
print(temp[:-1])
```
