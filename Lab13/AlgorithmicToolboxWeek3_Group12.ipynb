## 3-1: Money Change


```python
n = int(input())
count = 0

for i in [10, 5, 1]:
    if n >= i:
        q = n // i
        count += q
        n = n % i
        
        if n == 0:
            print(count)
```

## 3-2: Maximum Value of the Loot (Fractional Knapsack)


```python
n, W = [int(i) for i in input().split()]
lst = []

if W == 0:
    print(0)

for _ in range(n):
    v, w = [int(i) for i in input().split()]
    
    if v == 0:
        continue

    lst.append((v, w))

lst.sort(key=lambda x: x[0] / x[1], reverse=True)

total_value = 0

for v, w in lst:
    if W == 0:
        print(total_value)

    amt = min(w, W)
    total_value += amt * v / w
    W -= amt

print(total_value)
```

## 3-3: Car Fueling


```python
d = int(input())
m = int(input())
n = int(input())
i = input()
x = [0] + list(map(int, i.split())) + [d]

def MinRefill(x, m, n):
    current = 0
    count = 0

    while current <= n:
        last = current

        while current <= n and x[current + 1] - x[last] <= m:
            current = current + 1

        if current == last:
            return -1

        if current <= n:
            count = count + 1
    
    return count

print(MinRefill(x, m, n))
```

## 3-4: Maximum Advertisement Revenue (Maximum Dot Product)


```python
n = int(input())
ipt = input()
an = list(map(int, ipt.split()))
ipt = input()
bn = list(map(int, ipt.split()))

an.sort()
bn.sort()
cn = [an[i] * bn[i] for i in range(n)]

print(sum(cn))
```

## 3-5: Collecting Signatures (Covering Segments by Points)


```python
n = int(input())
lst = []

for _ in range(n):
    a, b = [int(i) for i in input().split()]
    lst.append((a, b))

lst.sort(key=lambda x: x[1])

index = 0
coordinates = []

while index < n:
    curr = lst[index]
    
    while index < n - 1 and curr[1] >= lst[index + 1][0]:
        index += 1
    coordinates.append(curr[1])
    index += 1

print(len(coordinates))
print(' '.join([str(i) for i in coordinates]))
```

## 3-6: Maximum Number of Prizes (Different Summands)


```python
n = int(input())
prize = []
i = 0

while n > 0:
    i = i + 1
    if n - i <= i:
        i = n
    prize.append(i)
    n = n - i

print(len(prize))

for x in prize:
    print(x, end=' ')
```

## 3-7: Maximum Salary (Largest Number)


```python
n = int(input())
digits = list(input().split())


def max_num(lst):
    n = len(lst)
    for i in range(n - 1):
        for i in range(n - 1 - i):
            if lst[i] + lst[i + 1] < lst[i + 1] + lst[i]:
                lst[i], lst[i + 1] = lst[i + 1], lst[i]
    return lst


digits = max_num(digits)
result = ''

for digit in digits:
    result = result + digit

print(result)
```
